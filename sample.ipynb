```python
import os
pth = 'data'
img_files =[]
labels = []
from sklearn.preprocessing import LabelEncoder
```

```python
for file_name in os.listdir(os.path.join(pth,'train')):
            if file_name.endswith(".png"):
                img_files.append(os.path.join(pth, 'train', file_name))
                labels.append(file_name.split("_")[1].split('.')[0])
```

```python
file_name
```

```python
targets = [[c for c in x] for x in labels]
targets_flat = [c for clist in targets for c in clist]
```

```python
l_enc = LabelEncoder()
l_enc.fit(targets_flat)
targets_enc = [l_enc.transform(x) for x in targets]
```
